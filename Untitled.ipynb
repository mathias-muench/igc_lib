```python
from igc_lib import Flight
from pdflight import flight_to_dataframe
flight = Flight.create_from_file("./575_13.igc")                                                                                                                                               
```

```python
df = flight_to_dataframe(flight)                                                                                                                                                              
df
```

```python
df.info()
```

```python
import folium                                                                                                                                                                                 
                                                                                                                                                                                              
# Create base map centered on flight                                                                                                                                                          
m = folium.Map(                                                                                                                                                                               
    location=[df['lat'].mean(), df['lon'].mean()],                                                                                                                                            
    zoom_start=12,                                                                                                                                                                            
    tiles='OpenStreetMap'  # or 'Stamen Terrain', 'CartoDB positron'                                                                                                                          
)                                                                                                                                                                                             
                                                                                                                                                                                              
# Add flight path                                                                                                                                                                             
folium.PolyLine(                                                                                                                                                                              
    locations=df[['lat', 'lon']].values,                                                                                                                                                      
    color='red',                                                                                                                                                                              
    weight=2.5,                                                                                                                                                                               
    opacity=1                                                                                                                                                                                 
).add_to(m)                                                                                                                                                                                   
                                                                                                                                                                                              
# Add takeoff/landing markers                                                                                                                                                                 
folium.Marker(                                                                                                                                                                                
    location=df.iloc[0][['lat', 'lon']].values,                                                                                                                                               
    popup='Takeoff',                                                                                                                                                                          
    icon=folium.Icon(color='green')                                                                                                                                                           
).add_to(m)                                                                                                                                                                                   
                                                                                                                                                                                              
folium.Marker(                                                                                                                                                                                
    location=df.iloc[-1][['lat', 'lon']].values,                                                                                                                                              
    popup='Landing',                                                                                                                                                                          
    icon=folium.Icon(color='red')                                                                                                                                                             
).add_to(m)                                                                                                                                                                                   
                                                                                                                                                                                              
# Display                                                                                                                                                                                     
m  
```

```python
import matplotlib.pyplot as plt        
import pandas as pd

# Identify transitions into circling mode                                                                                                                                                 
circling_starts = df[df['circling'] & (~df['circling'].shift(1, fill_value=False))]                                                                                                       
                                                                                                                                                                                          
# Plot ground speed for the first `max_seconds` after each circling start                                                                                                                 
plt.figure(figsize=(10, 6))                                                                                                                                                               
for idx, start_time in enumerate(circling_starts.index[10:20]):                                                                                                                                  
    # Get the segment of data for the next `max_seconds` after circling starts                                                                                                            
    segment = df.loc[start_time - pd.Timedelta(seconds=60) : start_time + pd.Timedelta(seconds=30)]                                                                                                         
    if len(segment) > 0:                                                                                                                                                                  
        plt.plot(                                                                                                                                                                         
            (segment.index - start_time).total_seconds(),  # Time since circling start                                                                                                    
            segment['gsp'],                                                                                                                                                               
            label=f'Circling Segment {idx + 1}',                                                                                                                                          
            marker='o',                                                                                                                                                                   
            linestyle='-',                                                                                                                                                                
            markersize=3                                                                                                                                                                  
        )                                                                                                                                                                                 
                                                                                                                                                                                          
plt.xlabel('Time since circling start (seconds)')                                                                                                                                         
plt.ylabel('Ground Speed (km/h)')                                                                                                                                                         
plt.title('Ground Speed in First 30 Seconds of Circling')                                                                                                                                 
plt.grid(True)                                                                                                                                                                            
plt.legend()                                                                                                                                                                              
plt.show()   
```
