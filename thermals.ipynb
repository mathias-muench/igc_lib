```python
import pandas as pd
from igc_lib import Flight
from pdflight import flight_to_dataframe,thermals_to_dataframe
flight = Flight.create_from_file("/home/muenchm/thinclient_drives/C:/Archive/muenchm/Data/IGC/551VQ6P1.igc")
#flight = Flight.create_from_file("/home/muenchm/thinclient_drives/C:/Users/muenchm/Desktop/2025-05-01 Wolfgang Janowitsch 562460.igc")
```

```python
fl = flight_to_dataframe(flight)                                                                                                                                                              
th = thermals_to_dataframe(flight)[1:]
th
```

```python
df = pd.DataFrame()
df["t"] = th.index + th
for i in th.index:
    df.loc[i, "v"] = (fl.loc[i + th.loc[i]].alt - fl.loc[i].alt) / th.loc[i].total_seconds()
    df.loc[i, "n"] = (fl.loc[i + pd.Timedelta(seconds=40)].alt - fl.loc[i].alt) / 40
    df.loc[i, "x"] = (fl.loc[i + th.loc[i] - pd.Timedelta(seconds=40)].alt - fl.loc[i + th.loc[i]].alt) / -40
#    th.loc[i, "s"] = fl.loc[i : i + th.loc[i].duration].gsp.std()
#    th.loc[i, "a"] = fl.loc[i : i + th.loc[i].duration].gsp.mean()
df["dn"] = df.n / df.v
df["dx"] = df.x / df.v
df
```
